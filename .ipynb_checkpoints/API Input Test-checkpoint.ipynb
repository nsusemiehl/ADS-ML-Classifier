## arXiv

```python
start_date = datetime(2023, 4, 18)
end_date = datetime(2023, 4, 20)

start = start_date.strftime("%Y%m%d")
end = end_date.strftime("%Y%m%d")
range_str = f"[{start}000000+TO+{end}000000]"
range_query = f"lastUpdatedDate:{range_str}"
# range_query = f"submittedDate:{range_str}"

subcat = ["GA", "CO", "EP", "HE", "IM", "SR"]

cat_query = "%28"  # open parenthesis
for n, s in enumerate(subcat):
    cat_query += f"astro-ph.{s}"
    if n < len(subcat)-1:
        cat_query += "+OR+"
    else:
        cat_query += "%29"  # close parenthesis
        
base_url = "http://export.arxiv.org/api/query?"
sort_query = f"max_results={999}"

full_query = f"search_query={cat_query}+AND+{range_query}&{sort_query}"

url = base_url + full_query
print(url)
```

```python
print(url)
response = urlopen(url).read()
feed = feedparser.parse(response)
# print(feed)
api_arxiv_titles = [e.title_detail["value"] for e in feed.entries]
api_arxiv_ids = [e.id.split("/")[-1] for e in feed.entries]
api_arxiv_ids = [re.sub("v\d", "", id_) for id_ in api_arxiv_ids]

print(len(api_arxiv_titles))
```

```python
n = 0
for title in api_arxiv_titles:
    print(title)
    n += 1
        
print(n)
```
