```python
checked_nonexoplanet_negative_bibcodes = []
with open('.//data//checked_nonexoplanet_negatives.txt', 'r') as f:
    for line in f:
        checked_nonexoplanet_negative_bibcodes.append(line.strip())
        
print(len(checked_nonexoplanet_negative_bibcodes))
```

```python
unchecked_bib_inds = []
for i in range(len(completed_non_ea_bibs)):
    if completed_non_ea_bibs[i] not in checked_nonexoplanet_negative_bibcodes:
        unchecked_bib_inds.append(i)
        
print(len(unchecked_bib_inds))

completed_non_ea_bibs = np.delete(completed_non_ea_bibs, unchecked_bib_inds)
completed_non_ea_text = np.delete(completed_non_ea_text, unchecked_bib_inds)

len(completed_non_ea_bibs)
```

```python
completed_non_ea_years = []
for i in completed_non_ea_bibs:
    completed_non_ea_years.append(int(i[:4]))
```

```python
plt.hist(completed_non_ea_years, bins=20)
plt.title("Non-EA Papers")
plt.xlabel("Year")
plt.show()
# are bad papers from certain years?
```
